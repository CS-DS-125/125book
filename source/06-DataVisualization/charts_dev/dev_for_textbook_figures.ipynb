```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# For nicer-looking charts
%config InlineBackend.figure_format = 'svg'
```

```python
df = pd.read_csv("broadway.csv", parse_dates=["Full"])
```

```python
df_grptypesum = df.groupby("Type").sum().reset_index()
```

```python
plt.bar("Type", "Gross", data=df_grptypesum)
plt.ylabel("Total Gross Revenue")
plt.title("Total Gross Revenue by Show Type")
def fmtr(x, pos):
    return f"${x/1000000000:.0f} billion"
plt.gca().yaxis.set_major_formatter(mpl.ticker.FuncFormatter(fmtr))
plt.show()
```

```python
df_grptypemean = df.groupby("Type").mean().reset_index()
```

```python
plt.bar("Type", "Gross", data=df_grptypemean)
plt.ylabel("Average Weekly Gross Revenue")
plt.title("Average Weekly Gross Revenue by Show Type")
def fmtr(x, pos):
    return f"${x:,.0f}"
plt.gca().yaxis.set_major_formatter(mpl.ticker.FuncFormatter(fmtr))
plt.show()
```

```python
plt.rcParams['figure.figsize'] = [10, 5]

df_Shubert = df.query("Theatre == 'Shubert'")
plt.plot("Full","Attendance", data=df_Shubert, linewidth=1)
plt.ylabel("Weekly Attendance")
plt.xlabel("Date")
plt.title("Weekly Attendance, Shubert Theatre")
plt.show()
```

```python
plt.rcParams['figure.figsize'] = [10, 5]

df_Minskoff = df.query("Theatre == 'Minskoff'")
plt.plot("Full","Attendance", data=df_Minskoff, linewidth=1)
plt.ylabel("Weekly Attendance")
plt.xlabel("Date")
plt.title("Weekly Attendance, Minskoff Theatre")
plt.show()
```

```python
plt.rcParams['figure.figsize'] = [10, 5]

df_Minskoff = df.query("Theatre == 'Minskoff'")
plt.plot("Full","Gross", data=df_Minskoff, linewidth=1)
plt.ylabel("Weekly Gross Revenue")
plt.xlabel("Date")
def fmtr(x, pos):
    return f"${x:,.0f}"
plt.gca().yaxis.set_major_formatter(mpl.ticker.FuncFormatter(fmtr))
plt.title("Weekly Gross Revenue, Minskoff Theatre")
plt.show()
```

```python
%config InlineBackend.figure_format = 'png'
plt.rcParams['figure.figsize'] = [10, 5]
plt.rcParams['figure.dpi'] = 200

plt.scatter("Attendance", "Gross", data=df, alpha=0.15, s=1)
plt.xlabel("Weekly Attendance")
plt.ylabel("Weekly Gross Revenue")
plt.gca().yaxis.set_major_formatter(mpl.ticker.FuncFormatter(fmtr))
plt.title("Revenue vs Attendance")
plt.show()
```

```python
%config InlineBackend.figure_format = 'png'
plt.rcParams['figure.figsize'] = [10, 5]
plt.rcParams['figure.dpi'] = 200

plt.scatter("Attendance", "Gross", c=df['Full'].map(lambda x: x.year), data=df, s=0.05)
plt.xlabel("Weekly Attendance")
plt.ylabel("Weekly Gross Revenue")
plt.gca().yaxis.set_major_formatter(mpl.ticker.FuncFormatter(fmtr))
plt.title("Revenue vs Attendance")
cbar = plt.colorbar()
plt.show()
```

```python
%config InlineBackend.figure_format = 'svg'

plt.hist("Attendance", data=df, bins="auto")
plt.xlabel("Attendance")
plt.ylabel("Frequency")
plt.title("Histogram of Weekly Attendance")
plt.show()
```
